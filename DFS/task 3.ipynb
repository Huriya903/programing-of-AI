```python
def dfs_water_jug(capA, capB, goal):
    stack = [((0, 0), [])]  
    visited = set()

    while stack:
        (a, b), path = stack.pop()

        if (a, b) in visited:
            continue
        visited.add((a, b))

        path = path + [(a, b)]

        if a == goal or b == goal:
            print("Solution Found!\n")
            for i in range(len(path)-1):
                curr = path[i]
                next_state = path[i+1]

                print(f"Step {i}: {curr} -> {next_state}")
            print(f"Final Step: {path[-1]}")
            return

        next = []
        next.append(((capA, b), "Fill Jug A"))
        next.append(((a, capB), "Fill Jug B"))
        next.append(((0, b), "Empty Jug A"))
        next.append(((a, 0), "Empty Jug B"))

        pour = min(a, capB - b)
        next.append(((a - pour, b + pour), "Pour A -> B"))
        pour = min(b, capA - a)
        next.append(((a + pour, b - pour), "Pour B -> A"))

        for state, rule in next:
            if state not in visited:
                stack.append((state, path))

    print("No Solution Found")

dfs_water_jug(4, 3, 2)
```
